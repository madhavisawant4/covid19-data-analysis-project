10.compare the number of cases increasing between the two states with line plot

```python
delhi=df[df["States"]=="Delhi"]
delhi.head()
#1st case found in delhi is on 2020-03-02
```

```python
delhi.tail()
#as of 2020-08-06 delhi has 140232  confirmed cases
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Date",y="Total_Confirmed_cases",data=delhi)
plt.legend(['Delhi'])
plt.show()
```

```python
goa=df[df["States"]=="Goa"]
goa.head()
#1st case found in goa is on 2020-03-26
```

```python
goa.tail()
#as of 2020-08-06 goa has 7423  confirmed cases
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Date",y="Total_Confirmed_cases",data=goa)
plt.legend(["Goa"])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Date",y="Total_Confirmed_cases",data=delhi)
sns.lineplot(x="Date",y="Total_Confirmed_cases",data=goa)
plt.legend(["Delhi","Goa"])
plt.show()
```

```python
maharashtra=df[df["States"]=="Maharashtra"]
maharashtra.head()
#1st case found in Maharashtra is on 2020-03-09
```

```python
maharashtra.tail()
#as of 2020-08-06 Maharashtra has 468265  confirmed cases
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Date",y="Total_Confirmed_cases",data=maharashtra)
plt.legend(["Maharashtra"])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Date",y="Total_Confirmed_cases",data=delhi)
sns.lineplot(x="Date",y="Total_Confirmed_cases",data=goa)
sns.lineplot(x="Date",y="Total_Confirmed_cases",data=maharashtra)
plt.legend(["Delhi","Goa","Maharashtra"])
plt.show()
#goa has least no. of cases then delhi and then maharashtra
```

11.With scatterplot show the number of cases and recovery rate of covid-19 patients

```python
x=df['New cases']
y=df['New recovered']
plt.figure(figsize=(10,7))
plt.scatter(x,y,color='g',marker='^',s=100)
plt.title("covid 19 data")
plt.xlabel("New cases")
plt.ylabel("New recovered")
plt.legend([["New cases","New recovered"]])

plt.show()
#cases less than 2500 the recovery rate is less but as no increases the more peoples have recovered as well
```

12.which are the top 5 states which have highest recovery rate

```python
merged=new_df1.groupby("States").sum().sort_values(by="Cured",ascending=False).head()
merged
```

```python
plt.figure(figsize=(10,7))
sns.lineplot(x="States",y="Cured",data=merged)
plt.show()
```
